```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as pr
```

## Scrapping Website

```python
url = "https://www.flipkart.com/search?q=pendrive&sid=6bo%2Cjdy%2Cuar&as=on&as-show=on&otracker=AS_QueryStore_HistoryAutoSuggest_1_3_na_na_na&otracker1=AS_QueryStore_HistoryAutoSuggest_1_3_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=pendrive%7CPen+Drives&requestId=abec14ad-e3c2-465f-922f-dc817203fdd9&as-searchtext=pen"
```

## Wesbite Response

```python
res = requests.get(url)
```

```python
res
```

## parsing

```python
soup = BeautifulSoup(res.text)
```

```python
soup
```

## Pendrive name

```python
pendrive = soup.select(".s1Q9rs")
```

```python
pendrive
```

```python
prices=soup.select('div~ div+ div ._4ddWXP ._30jeq3 , ._13oc-S div:nth-child(1) ._4ddWXP ._30jeq3')
```

```python
prices
```

```python
price_title=[]
```

```python
for item in prices:
    price_title.append(item.text)
```

```python
price_title
```

```python
pendrive_names=[]
```

```python
for item in pendrive:
    pendrive_names.append(item.text)
```

```python
pendrive_names
```

## Data Frame

```python
df=pr.DataFrame(pendrive_names,columns=['pendrive names'])
```

```python
df
```

```python
df['price_title']=price_title
```

```python
df
```

```python
df.to_csv('pk.csv')
```
